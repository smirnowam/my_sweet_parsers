```python
#сосздаем ссылку на пул ликвидности в revert.finance из их API
revert = "https://staging-api.revert.finance/v1/positions/%s" #"https://revert.finance/#/uniswap-position/%s"
blck = ["optimism/uniswapv3/","polygon/uniswapv3/","arbitrum/uniswapv3/","bnb/uniswapv3/"]

#УКАЗЫВАЕМ id pool'ов:

#id pool для optimism:
id_pool_op = [401424,400777,400784,400967]
#id pool для polygon:
id_pool_polygon = []
#id pool для arbitrum:
id_pool_arb = [52435]
#id pool для BNB:
id_pool_bnb = [41239]

#делаем из каждого id пула текст
for i in range(len(id_pool_op)):
    id_pool_op[i] = str(id_pool_op[i])
for i in range(len(id_pool_polygon)):
    id_pool_polygon[i] = str(id_pool_polygon[i])
for i in range(len(id_pool_arb)):
    id_pool_arb[i] = str(id_pool_arb[i])
for i in range(len(id_pool_bnb)):
    id_pool_bnb[i] = str(id_pool_bnb[i])
    
#прибавляем к № пула часть ссылки, которая приведет к нужному блокчейну:
pool = []
for n in range(len(blck)):
    if blck[n] == "optimism/uniswapv3/":
        for o in range(len(id_pool_op)):
            pool.append(blck[n]+id_pool_op[o])
    elif blck[n] == "polygon/uniswapv3/":
        for p in range(len(id_pool_polygon)):
            pool.append(blck[n]+id_pool_polygon[p])
    elif blck[n] == "arbitrum/uniswapv3/":
        for a in range(len(id_pool_arb)):
            pool.append(blck[n]+id_pool_arb[a])
    elif blck[n] == "bnb/uniswapv3/":
        for bb in range(len(id_pool_bnb)):
            pool.append(blck[n]+id_pool_bnb[bb])
    else:
        None

#делаем ссылочки:
links = []
for i in range(len(pool)):
    links.append(revert%pool[i])
    
#делаем ссылочки для финальной таблицы:
pool_numbers = list(id_pool_op+id_pool_polygon+id_pool_arb+id_pool_bnb)
blckcn = ["optimism/","polygon/","arbitrum/","bnb/"]
rvrt = "https://revert.finance/#/uniswap-position/%s"
pol = []
for z in range(len(blckcn)):
    if blckcn[z] == "optimism/":
        for y in range(len(id_pool_op)):
            pol.append(blckcn[z]+id_pool_op[y])
    elif blckcn[z] == "polygon/":
        for u in range(len(id_pool_polygon)):
            pol.append(blckcn[z]+id_pool_polygon[u])
    elif blckcn[z] == "arbitrum/":
        for t in range(len(id_pool_arb)):
            pol.append(blckcn[z]+id_pool_arb[t])
    elif blckcn[z] == "bnb/":
        for q in range(len(id_pool_bnb)):
            pol.append(blckcn[z]+id_pool_bnb[q])
    else:
        None
urls = []
for d in range(len(pol)):
    urls.append(rvrt%pol[d])

#делаем запросы:
soup = []
for link in links:
    response = requests.get(link)
    if response.status_code == 200:
        soup.append(BeautifulSoup(response.content, 'html.parser'))
    else:
        None
        
#переводим запрос в удобоваримый формат:
sp=[]
data = []
for s in range(len(soup)):
    sp.append(str(soup[s]))
for d in range(len(sp)):
    data.append(json.loads(sp[d]))
    
#переменные для заполнения:
total_fees0, total_fees1, fees_value, in_range, age, collected_fees0, collected_fees1, uncollected_fees0,uncollected_fees1, current_amount0, current_amount1, diffs0, diffs1, total_deposits0, total_deposits1, total_withdrawn0, total_withdrawn1, network, exchange, pool_price, price_upper, price_lower, ts, pnl, apr, fee_apr, token0_price1, token1_price1, token0, token1, token0_price2, token1_price2, token0_id, token1_id= ([],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[])

#заполняем переменные:
for k in data:
    inner_data = k.get("data",{})
    value1 = inner_data.get("total_withdrawn0",[])
    value2 = inner_data.get("total_withdrawn1",[])
    value3 = inner_data.get("total_fees0",[])
    value4 = inner_data.get("total_fees1",[])
    value5 = inner_data.get("fees_value",[])
    value6 = inner_data.get("in_range",[])
    value7 = inner_data.get("age",[])
    value8 = inner_data.get("collected_fees0",[])
    value9 = inner_data.get("collected_fees1",[])
    value10 = inner_data.get("uncollected_fees0",[])
    value11 = inner_data.get("uncollected_fees1",[])
    value12 = inner_data.get("current_amount0",[])
    value13 = inner_data.get("current_amount1",[])
    value14 = inner_data.get("diffs0",[])
    value15 = inner_data.get("diffs1",[])
    value16 = inner_data.get("total_deposits0",[])
    value17 = inner_data.get("total_deposits1",[])
    value18 = inner_data.get("network",[])
    value19 = inner_data.get("exchange",[])
    value20 = inner_data.get("pool_price",[])
    value21 = inner_data.get("price_upper",[])
    value22 = inner_data.get("price_lower",[])
    value23 = inner_data.get("history_24h",[])
    value24 = inner_data.get("cash_flows",[])
    value25 = inner_data.get("tokens",[])
    try:
        total_withdrawn0.append(value1)
    except (IndexError, KeyError):
        total_withdrawn0.append(None)
    try:
        total_withdrawn1.append(value2)
    except (IndexError, KeyError):
        total_withdrawn1.append(None)
    try:
        total_fees0.append(value3)
    except (IndexError, KeyError):
        total_fees0.append(None)
    try:
        total_fees1.append(value4)
    except (IndexError, KeyError):
        total_fees1.append(None)
    try:
        fees_value.append(value5)
    except (IndexError, KeyError):
        fees_value.append(None)
    try:
        in_range.append(value6)
    except (IndexError, KeyError):
        in_range.append(None)
    try:
        age.append(value7)
    except (IndexError, KeyError):
        age.append(None)
    try:
        collected_fees0.append(value8)
    except (IndexError, KeyError):
        collected_fees0.append(None)
    try:
        collected_fees1.append(value9)
    except (IndexError, KeyError):
        collected_fees1.append(None)
    try:
        uncollected_fees0.append(value10)
    except (IndexError, KeyError):
        uncollected_fees0.append(None)
    try:
        uncollected_fees1.append(value11)
    except (IndexError, KeyError):
        uncollected_fees1.append(None)
    try:
        current_amount0.append(value12)
    except (IndexError, KeyError):
        current_amount0.append(None)
    try:
        current_amount1.append(value13)
    except (IndexError, KeyError):
        current_amount1.append(None)
    try:
        diffs0.append(value14)
    except (IndexError, KeyError):
        diffs0.append(None)
    try:
        diffs1.append(value15)
    except (IndexError, KeyError):
        diffs1.append(None)
    try:
        total_deposits0.append(value16)
    except (IndexError, KeyError):
        total_deposits0.append(None)
    try:
        total_deposits1.append(value17)
    except (IndexError, KeyError):
        total_deposits1.append(None)
    try:
        network.append(value18)
    except (IndexError, KeyError):
        network.append(None)
    try:
        exchange.append(value19)
    except (IndexError, KeyError):
        exchange.append(None)
    try:
        pool_price.append(value20)
    except (IndexError, KeyError):
        pool_price.append(None)
    try:
        price_upper.append(value21)
    except (IndexError, KeyError):
        price_upper.append(None)
    try:
        price_lower.append(value22)
    except (IndexError, KeyError):
        price_lower.append(None)
    try:
        ts.append(value23[0].get("ts",{}))
    except (IndexError, KeyError):
        ts.append(None)
    try:
        pnl.append(value23[0].get("pnl",{}))
    except (IndexError, KeyError):
        pnl.append(None)
    try:
        apr.append(value23[0].get("apr",{}))
    except (IndexError, KeyError):
        apr.append(None)
    try:
        fee_apr.append(value23[0].get("fee_apr",{}))
    except (IndexError, KeyError):
        fee_apr.append(None)
    try:
        token0_price1.append(value24[0].get("prices",{}).get("token0",{})["usd"])
    except (IndexError, KeyError):
        token0_price1.append(None)
    try:
        token1_price1.append(value24[0].get("prices",{}).get("token1",{})["usd"])
    except (IndexError, KeyError):
        token1_price1.append(None)
    try:
        token0_id.append(list(value25)[0])
    except (IndexError, KeyError):
        token0_id.append(None)
    try:
        if isinstance(value25, dict):
            symbols = list(value25.values())
            token0.append(symbols[0].get('symbol'))
        else:
            token0.append(None)
    except (IndexError, KeyError):
        token0.append(None)
    try:
        if isinstance(value25, dict):
            price_t0 = list(value25.values())
            token0_price2.append(price_t0[0].get('price'))
        else:
            token0_price2.append(None)
    except (IndexError, KeyError):
        token0_price2.append(None)
    #try:
    #    price_t0 = [item['price'] for item in value25.values()]
    #    token0_price2.append(price_t0[0])
    #    #token0_price2.append(list(value25.items())[0][1]["price"])
    #except (IndexError, KeyError):
    #    token0_price2.append(None)
    try:
        token1_id.append(list(value25)[1])
    except (IndexError, KeyError):
        token1_id.append(None)
    try:
        if isinstance(value25, dict):
            symbols = list(value25.values())
            token1.append(symbols[1].get('symbol'))
            #token1.append(list(value25.items())[1][1]["symbol"])
        else:
            token1.append(None)
        #symbols = [item['symbol'] for item in value25.values()]
        #token1.append(symbols[1])
        #token1.append(list(value25.items())[1][1]["symbol"])
    except (IndexError, KeyError):
        token1.append(None)
    try:
        if isinstance(value25, dict):
            price_t1 = list(value25.values())
            token1_price2.append(price_t1[1].get('price'))
        else:
            token1_price2.append(None)
    except (IndexError, KeyError):
        token1_price2.append(None)
    #try:
    #    token1_price2.append(list(value25.items())[1][1]["price"])
    #except (IndexError, KeyError):
    #    token1_price2.append(None)

tabl = {
    'Pool Number': pool_numbers,
    'Pool Link': urls,
    'network': network,
    'exchange': exchange,
    'in_range': in_range,
    'age_days': age,
    'pool_price': pool_price,
    'price_upper': price_upper,
    'price_lower': price_lower,
    'total_deposits0': total_deposits0,
    'total_deposits1': total_deposits1,
    'ts': ts,
    'pnl': pnl,
    'apr': apr,
    'fee_apr': fee_apr,
    'token0': token0,
    'token1': token1,
    'token0_price2': token0_price2,
    'token1_price2': token1_price2,
    'current_amount0': current_amount0,
    'current_amount1': current_amount1,
    'total_withdrawn_token0': total_withdrawn0,
    'total_withdrawn_token1': total_withdrawn1,
    'total_fees_token0': total_fees0,
    'total_fees_token1': total_fees1,
    'collected_fees0': collected_fees0,
    'collected_fees1': collected_fees1,
    'uncollected_fees0': uncollected_fees0,
    'uncollected_fees1': uncollected_fees1,
    'fees_value': fees_value,
    'diffs0': diffs0,
    'diffs1': diffs1,
    'token0_price1': token0_price1,
    'token1_price1': token1_price1,
    'token0_id': token0_id,
    'token1_id': token1_id
        }    
#ебём!
df = pd.DataFrame(tabl)
df.to_excel("нихренасебе.xlsx",
             sheet_name='нихренасебе') 
```
